```python
import cvxpy as cp

# BIC Site: x1 (standard), x2 (deluxe)
x1 = cp.Variable(integer=True)
x2 = cp.Variable(integer=True)


# Define the objective function
objective = cp.Maximize(10*x1  + 15*x2)

# Define the constraints
constraints = [
    4*x1 + 2*x2 <= 80,
    2*x1 + 5*x2 <= 60,
    4*(x1 + x2) <= 120,
    x1 >= 0,
    x2 >= 0]

# Define the optimization problem
problem = cp.Problem(objective, constraints)

# Solve the problem
problem.solve()

# Print the status of the solution
print("Status:", problem.status)

# Print the optimal values of the variables
print("x1 = {:.2f}".format(x1.value))
print("x2 = {:.2f}".format(x2.value))


# Define the profit per unit for each product
profit_per_unit = {
    'x1': 10,
    'x2': 15
}

# Define the optimal values of the variables
optimal_values = {
    'x1': x1.value,
    'x2': x2.value
}


# Calculate the total profit
total_profit = sum([profit_per_unit[var] * optimal_values[var] for var in optimal_values])

print("Total Profit:", total_profit)
```

3. Now, write the optimization model for the joint optimization of both sites, and provide the optimal solution and profit levels.

```python
import cvxpy as cp

# BIC Site: x1 (standard), x2 (deluxe)
# CIB Site:y1 (standard), y2 (deluxe)
# Decision variables
x1 = cp.Variable(integer=True)
x2 = cp.Variable(integer=True)
y1 = cp.Variable(integer=True)
y2 = cp.Variable(integer=True)

# Define the objective function
objective = cp.Maximize(10*(x1 + y1) + 15*(x2 + y2))

# Define the constraints
constraints = [
    4*x1 + 2*x2 <= 80,
    2*x1 + 5*x2 <= 60,
    4*(x1 + x2) <= 75,
    5*y1 + 3*y2 <= 60,
    5*y1 + 6*y2 <= 75,
    4*(y1 + y2) <= 45,
    4*(x1 + x2 + y1 + y2) <= 120,
    x1 >= 0,
    x2 >= 0,
    y1 >= 0,
    y2 >= 0
]

# Define the optimization problem
problem = cp.Problem(objective, constraints)

# Solve the problem
problem.solve()

# Print the status of the solution
print("Status:", problem.status)

# Print the optimal values of the variables
print("x1 = {:.2f}".format(x1.value))
print("x2 = {:.2f}".format(x2.value))
print("y1 = {:.2f}".format(y1.value))
print("y2 = {:.2f}".format(y2.value))

# Define the profit per unit for each product
profit_per_unit = {
    'x1': 10,
    'x2': 15,
    'y1': 10,
    'y2': 15
}

# Define the optimal values of the variables
optimal_values = {
    'x1': x1.value,
    'x2': x2.value,
    'y1': y1.value,
    'y2': y2.value
}

# Calculate the total profit
total_profit = sum([profit_per_unit[var] * optimal_values[var] for var in optimal_values])

print("Total Profit:", total_profit)

```

```python
def purchase_hotels(df, budget):
    # Sort hotels by profitability in descending order
    sorted_hotels = df.sort_values(by='Profitability', ascending=False)
    
    purchased_hotels = []
    total_spent = 0
    total_profitability = 0

    for _, hotel in sorted_hotels.iterrows():
        if total_spent + hotel['Price'] <= budget:
            purchased_hotels.append(hotel)
            total_spent += hotel['Price']
            total_profitability += hotel['Profitability']

    # Calculate average profitability of purchased hotels
    if purchased_hotels:
        average_profitability = total_profitability / len(purchased_hotels)
    else:
        average_profitability = 0

    return purchased_hotels, total_profitability, average_profitability, total_spent

# Try with $10 million budget
hotels_10m, total_profit_10m, avg_profit_10m, total_spent_10m = purchase_hotels(df, 10000000)

# Try with $20 million budget
hotels_20m, total_profit_20m, avg_profit_20m, total_spent_20m = purchase_hotels(df, 20_000_000)

# Output the results
print("With $10 million budget:")
print("Total Predicted Profitability:", total_profit_10m)
print("Average Predicted Profitability:", avg_profit_10m)
print("Purchased Hotels:", [hotel['Hotel'] for hotel in hotels_10m])
print("Total Spent:", total_spent_10m)

print("\nWith $20 million budget:")
print("Total Predicted Profitability:", total_profit_20m)
print("Average Predicted Profitability:", avg_profit_20m)
print("Purchased Hotels:", [hotel['Hotel'] for hotel in hotels_20m])
print("Total Spent:", total_spent_20m)
```

```python
import pandas as pd
import numpy as np
import cvxpy as cvx

# Load data
data_url = 'https://raw.githubusercontent.com/ormarketing/OD/master/OD.csv'
df = pd.read_csv(data_url)

# Separate males and females
females = df[df['Gender'] == 'Female']
males = df[df['Gender'] == 'Male']

# Initialize variables
num_females = len(females)
num_males = len(males)
activities = ['Sports', 'Theater', 'Religion', 'Outdoors', 'Eating Out']

# Create compatibility score matrix
compatibility_matrix = np.zeros((num_females, num_males))

for i in range(num_females):
    for j in range(num_males):
        # Calculate the compatibility score
        compatibility_matrix[i, j] = (females.iloc[i][activities] * males.iloc[j][activities]).sum()


# Optimization variables
X = cvx.Variable((num_females, num_males), boolean=True)

# Objective: Maximize sum of compatibility scores
objective = cvx.Maximize(cvx.sum(cvx.multiply(X, compatibility_matrix)))

# Constraints
# Each female is matched exactly once
constraints = [cvx.sum(X, axis=1) == 1]

# Each male is matched at most once
constraints += [cvx.sum(X, axis=0) <= 1]

# Solve the problem
problem = cvx.Problem(objective, constraints)
problem.solve()

# Output results
print("The maximum aggregate match score is:", problem.value)
matched_pairs = np.argwhere(X.value > 0.9)
for index in matched_pairs:
    female_name = females.iloc[index[0]]['Unnamed: 0']
    male_name = males.iloc[index[1]]['Unnamed: 0']
    print(f"Match {female_name} with {male_name}")
```
